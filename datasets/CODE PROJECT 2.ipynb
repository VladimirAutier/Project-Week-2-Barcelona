# Do we have more male of female unemployed?

```python
data_gender = data.groupby(['Year','Month','Gender']).sum ()
data_gender.rename(dict,inplace=True)
data_gender.sort_values(['Year','Month'],ascending = True,inplace=True)
data_gender
data_gender.drop([ 'District Code', 'Neighborhood Code'], axis=1,inplace=True)
data_gender
```
